# Python data analysis project

**The cereal dataset contains information about various cereals, including their names, types , nutritional values , minerals  vitamins, weight, cups per serving, and ratings. It provides an overview of the nutritional composition and characteristics of different cereal products.**

```python
#Import the essential libraries.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the cereal dataset.
```

```python
cereal_df=pd.read_csv('cereal.csv')
cereal_df
```

```python
cereal_df.info()
```

```python
#Printing the top 10 cereals from the dataset.
```

```python
cereal_df.head(10)
```

```python
#Check for any null values in the dataset if any.
```

```python
cereal_df.isnull()
```

**DATA CLEANING**

```python
#Dropping unnecessary columns.
```

```python
if 'mfr' in cereal_df.columns and 'shelf' in cereal_df.columns:
    cereal_df = cereal_df.drop(['mfr', 'shelf'], axis=1)
```

```python
cereal_df
```

```python
#dropping null values in the dataset if any.
```

```python
cereal_df=cereal_df.dropna()
```

```python
cereal_df
```

```python
#converting the datatype of type column.
```

```python
cereal_df['type']=cereal_df['type'].astype('category')
```

```python
cereal_df
```

```python
#changing the column names of column which depict a vague description.
new_column_name={'carbo':'carbohydrates','potass':'potassium'}
cereal_df=cereal_df.rename(columns=new_column_name)
```

```python
#checking for any duplicate values and removing them.
print(cereal_df.duplicated())
cereal_df.drop_duplicates(inplace=True)
```

**ANALYZING THE DATA WITH VARIOUS QUERIES**

```python
#Which cereal has the highest calorie content? 
```

```python
highest_calorie_cereal=cereal_df.loc[cereal_df["calories"].idxmax()]
print(f"The cereal with the highest calorie content is '{highest_calorie_cereal['name']}' with {highest_calorie_cereal['calories']} calories.")
```

```python
#visualize the top 10 cereals by calorie content.
```

```python
top_10_calories=cereal_df.nlargest(10,'calories')
plt.figure(figsize=(11,6))
sns.barplot(x="name",y="calories",data=top_10_calories)
plt.title('TOP 10 cereals by calorie content')
plt.xlabel('Cereal Name')
plt.ylabel('calories')
plt.show()
```

```python
#what is the average protein content across all cereals?
```

```python
avg_protein=cereal_df['protein'].mean()
print(f"the average protein content across all cereals is {avg_protein:.2f} grams.")
```

```python
# we will also visualize the distribution of protein content.
```

```python
plt.figure(figsize=(8,6))
sns.histplot(data=cereal_df,x="protein",bins=20)
plt.title('Distribution of protein content')
plt.xlabel('protein')
plt.ylabel('count')
plt.show()
```

```python
#Is there any difference in the nutritional value of hot and cold cereals?
```

```python
hot_cereals=cereal_df[cereal_df["type"]=='H']
cold_cereals=cereal_df[cereal_df["type"]=='C']
print("Mean values for hot cereals:")
print(hot_cereals[['calories','protein','fat','fiber','carbohydrates','sugars']].mean())
print("\nMean values for cold cereals:")
print(cold_cereals[['calories','protein','fat','fiber','carbohydrates','sugars']].mean())
```

```python
#visuaalize their difference in content
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(x='type',y='calories',data=cereal_df,palette='Set2')
plt.title('Calorie content by Cereal type')
plt.xlabel('Cereal Type')
plt.ylabel('Calories')
plt.show()
```

```python
#which cereals have the highest fibre content?
```

```python
high_fiber_cereals=cereal_df.sort_values(by='fiber',ascending=False)
print(high_fiber_cereals[['name','fiber']].head(10))
```

```python
#visualize the cereals with highest fibre content.
top_10_fiber=high_fiber_cereals.nlargest(10,'fiber')
plt.figure(figsize=(10,6))
sns.barplot(x="name",y="fiber",data=top_10_fiber,palette='viridis')
plt.title('top 10 cereals by fiber content')
plt.xlabel('cereal name')
plt.ylabel('fiber')
plt.show()
```

```python
#Is there any correlation between sugar content and rating?
```

```python
sugar_rating_corr=cereal_df['sugars'].corr(cereal_df['rating'])
print(f"the correlation between sugar content and rating is {sugar_rating_corr:.2f}.")
```

```python
#visualize the relationship between the sugar content and rating.
```

```python
plt.figure(figsize=(8,6))
sns.regplot(x='sugars',y='rating',data=cereal_df)
plt.title('Sugar content vs rating')
plt.xlabel('sugar')
plt.ylabel('rating')
plt.show()
```

**Conclusion:** ** after conducting a data analysis on the cereal dataset, we can draw various insights and conclusions. through this analysis we identified
cereals with highest and lowest calorie content that lets consumer make the right choice for them based on their preference.we also examined correlation between sugar content and the cereals rating.**
